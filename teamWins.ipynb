## Build team wins file

```python
# dependencies
import pandas as pd
```

```python
# consuming data file with years from 2014 and 2018
tradStats_df = pd.read_csv('Resources/winloss_by_teams.csv')
tradStats_df.head()
```

### Data Cleansing and Checks

```python
# check all columns and names
tradStats_df.columns
```

```python
removed1 = tradStats_df.replace({'NBA':''}, regex=True)
replace1 = removed1.replace({'BRO':'BRK'}, regex=True)
replace2 = replace1.replace({'CHA':'CHO'}, regex=True)
replace3 = replace2.replace({'OCT':'OKC'}, regex=True)
```

```python
tradStats_clean = replace3.loc[(replace3['Season'] >= '(2014-15)') & (replace3['Season'] <= '(2018-19)'),:]
tradStats_clean.head()
```

```python
tradStats_clean['Year'] = tradStats_clean['Season'].str[1:5]
tradStats_clean
```

```python
# checking data counts
tradStats_clean.count()
```

```python
# check NaN values before replacing it with 0
tradStats_clean.isnull().values.any()
```
